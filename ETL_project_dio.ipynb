Criar arquivo csv com texto livre

````python
import pandas as pd

dados_texto = [
    "João Silva, CPF 123.456.789-01, mora em São Paulo",
    "Maria Oliveira, CPF 987.654.321-02, reside no Rio de Janeiro",
    "Carlos Pereira, CPF 456.789.123-03, vive em Belo Horizonte",
    "Ana Souza, CPF 741.852.963-04, mora em Curitiba",
    "Pedro Santos, CPF 159.357.258-05, reside em Porto Alegre",
    "Juliana Lima, CPF 357.951.456-06, mora em Salvador",
    "Fernando Rocha, CPF 258.369.147-07, vive em Recife",
    "Patrícia Mendes, CPF 951.753.159-08, mora em Fortaleza",
    "Lucas Teixeira, CPF 159.456.753-09, reside em Goiânia",
]

df = pd.DataFrame({"texto_livre": dados_texto})
df.to_csv("dados_texto_livre.csv", index=False, encoding="utf-8")

print("CSV gerado com sucesso: dados_texto_livre.csv")
````

Extract

````python
free_text_df = pd.read_csv('dados_texto_livre.csv')
users_info = df['texto_livre'].tolist()

print("Extraçao realizada com sucesso" if len(users_info) == len(dados_texto) else f"Extraçao falhou, apenas {len(users_info)} usuarios carregados")
````

Transform

````python
import json
from openai import OpenAI
from google.colab import userdata


api_key = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=api_key)

def text_to_json(text:str):
    resposta = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Extraia os dados do texto em json, com as chaves: Nome, CPF e Cidade, o CPF deve conter apenas numeros."},
            {"role": "user", "content": text}
        ]
    )
    return resposta.choices[0].message.content.strip("```json")

formated_user_info = []

for text in users_info:
  user_text = text_to_json(text)
  user_dict = json.loads(user_text)
  formated_user_info.append(user_dict)
  print(user_dict)

print("Dados transformados com sucesso" if len(formated_user_info) == len(users_info) else f"Falha na transformaçao dos dados, apenas {len(formated_user_info)} dados foram transformados.")
````

Load

````python
print(len(formated_user_info))
df = pd.DataFrame(formated_user_info)
df.to_excel("dados.xlsx", sheet_name="Usuarios", index=False)
print(df.head(20))
````
